### 关联分析

参考:[Apriori算法详解](https://blog.csdn.net/zhihua_oba/article/details/73144606)

从大规模数据集中寻找物品间的隐含关系被称作**关联分析**或者**关联规则学习**。

关联规则中的常见概念:

- **项集**
假设$T={t_1,t_2,t_N}$是所有元素的集合，包含0个或者多个元素的集合就被称为项集，若项集中包含k个元素，则称它为**K-项集**。

- **频繁项集**
频繁项集是经常出现在一块儿的物品集合，定义大于等于**最小支持度阈值**的项集为频繁项集。

- **关联规则**
关联规则暗示两种物品之间可能存在的关系,比如A->B。

- **支持度**
支持度被定义为数据集中包含该项集的比例。

- **置信度**
置信度（又称为可信度）是针对关联规则定义，对于规则A->B的置信度被定义为$\frac{支持度(A,B)}{支持度(A)}$。

- **超集和子集**
若$A \subseteq B$,则A是B的子集，B是A的超集。

### Apriori算法原理

参考:[Apriori算法详解](https://www.jianshu.com/p/ff82fb98855d)

Apriori算法是一种针对大规模数据的关联规则学习算法。

Apriori算法依据以下两条定律:

- **定律1**：如果一个集合是频繁项集，则它的所有子集都是频繁项集。如果{A,B}是频繁的，即{A,B}出现的次数大于等于最小支持度阈值，则它的子集{A},{B}出现的次数必定大于最小支持度阈值，那么{A},{B}也一定是频繁的。

- **定律2**：如果一个集合不是频繁项集，则它的所有超集都不是频繁项集。假设如果{A}不是频繁项集，即A出现的次数小于最小支持度阈值，则它的任何超集如{A,B}出现的次数必定小于最小支持阈值，因此其超集必定也不是频繁项集。

APriori算法过程分为以下两步:

#### step1:寻找频繁项集

Apriori算法的两个输入参数分别是最小支持度和数据集。该算法首先会生成所有单个元素的项集列表。接着扫描数据集来查看哪些项集满足最小支持度要求，那些不满足最小支持度的集合会被去掉（剪枝）。然后，对剩下来的集合进行组合以生成包含两个元素的项集（连接）。接下来，再重新扫描交易记录，去掉不满足最小支持度的项集。该过程重复进行直到所有项集都被去掉。

最终所有项集的支持度都小于最小支持度阈值,算法终止。上图中绿色标志的项集即为频繁项集。


#### step2:生成关联规则

一旦找出所有频繁项集，就可以直接由他们产生强关联规则（**强关联规则满足最小支持度以及最小可信度**）。对于置信度，可用下式计算：
$$
confidence (A \rightarrow B)=P(B \mid A)=\frac{\text { support }(A \bigcup B)}{\text { support }(A)}
$$

关联规则按照如下规则产生:
* 对于每个频繁项集S，产生S的所有非空子集。
* 对于S的每个非空子集 $a$，如果 $\frac{\text {support}(S)}{\text {support}(a)} \geq \min _{-}$confidence$,$ 则输出规则” $a \rightarrow(S-a)$ ” 其
中，min_confidence为最小置信度阈值。

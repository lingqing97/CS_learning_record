### 前言

K近邻算法:
* 优点:精度高、对异常值不敏感、无数据输入假定
* 缺点:计算复杂度高、空间复杂度高
* 适用数据范围:数值型和标称型

### 实战:使用K-近邻算法改进约会网站的配对效果

在datingTestSet.txt文件中，每个样本数据占据一行，总共有1000行，每个样本包含以下3种特征:
* 每年获得的飞行常客里程数
* 玩视频游戏所耗时间百分比
* 每周消费的冰激凌公升数

每个样本划分为3种类型:
* 不喜欢的人（didntLike）
* 魅力一般的人（smallDoses）
* 极具魅力的人（largeDoses）

#### 读取数据

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#读取数据
dataset=pd.read_table("datasets/datingTestSet.txt",sep="\t")
X=dataset.iloc[:,:3].values
Y=dataset.iloc[:,3].values
```

```python
#将标签数据转为数值数据
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
Y=encoder.fit_transform(Y)
```

#### 分析数据

```python
#这里选取前两个属性进行可视化分析
plt.scatter(X[:,0],X[:,1],15.0*(Y+1),15.0*(Y+1))
plt.xlabel("Frequent flyer miles each year")
plt.ylabel('Play video games by a percentage')
plt.show()
```

#### 数据预处理

由于在K近邻算法中数字值最大的属性对计算结果的影响最大，比如:
$$
\sqrt{(0-67)^{2}+(20000-32000)^{2}+(1.1-0.1)^{2}}
$$
所以我们一般要对数据进行归一化处理。

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

#### 划分训练集和测试集

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=0)
```

#### 模型训练

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,Y_train)
```

#### 预测

```python
y_pred=classifier.predict(X_test)
print(y_pred)
```

#### 结果分析

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
print(cm)
```

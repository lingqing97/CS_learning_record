###  softmax回归模型

softmax回归的输出值个数等于标签里的类别数。假设输入的图像有4个像素分别记为$x_1$,$x_2$,$x_3$,$x_4$,真实标签为狗猫鸡三种标签，分别记为$y_1$,$y_2$,$y_3$.因为一共有4中特征和3中输出动物类别，所以权重包含12个标量，偏差包含3个标量，且对每个输入计算$o_1$,$o_2$,$o_3$这3个输出:


用神经网络表示，softmax回归的输出层也是一个全连接层。


分类问题需要得到离散的预测输出，一个简单的方法是将输出值$o_i$当做预测类别是i的置信度，并将值最大的输出所对应的类作为预测输出，集输出$argmax_io_i$.

由于输出值$o_i$的范围不确定，我们难以直观上判断这些值得意义。通过softmax运算法将输出值得范围统一可以解决这个问题。

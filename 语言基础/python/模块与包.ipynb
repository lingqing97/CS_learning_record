```python
#将代码组织封装成包


#要封装成包，必须确保每个目录都定义了一个`__init__.py``文件

# graphics/
#     __init__.py
#     primitive/
#         __init__.py
#         line.py
#         fill.py
#         text.py
#     formats/
#         __init__.py
#         png.py
#         jpg.py

#在模块中定义一个变量`__all__`来明确地导出需要列出的内容
#如果将`__all__`定义成一个空列表，没有东西将被导入。如果`__all__`包含未定义的名字，在导入时引起AttributeError

__all__=['spam','grok']

def spam():
    pass

def grok():
    pass

blah=42

# mypackage/
#     __init__.py
#     A/
#         __init__.py
#         spam.py
#         grok.py
#     B/
#         __init__.py
#         bar.py

#如果模块mypackage.A.spam要导入同目录下的模块grok,它应该包括的import语句如下:
from . import grok

#如果模块mypackage.A.spam要导入不同目录下的模块B.bar,它应该使用的import语句如下:
from ..B import bar


#将模块分割成多个文件

# mymodule/
#     __init__.py
#     a.py
#     b.py

#在a.py文件中插入如下代码:
# a.py
class A:
    def spam(self):
        print('A.spam')
        
# b.py
from .a import A
class B(A):
    def bar(self):
        print('B.bar')

# 在__init__.py中，将2个文件粘合在一起
# __init__.py
from .a import A
from .b import B

#之后可以将mymodule作为一个单一的逻辑模块

import mymodule
a=mymodule.A()
a.spam()
b=mymodule.B()
b.bar()
```
